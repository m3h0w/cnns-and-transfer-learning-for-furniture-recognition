```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
'''
img = Image.open("./data/train/1/89809.jpeg").convert('LA')
img2 = Image.open("./data/train/1/89810.jpeg").convert('LA')

img2.show()

pixels = img2.getdata() # get the pixels as a flattened sequence
white_thresh = 240
n_white = 0
for pixel in pixels:
    if pixel[0] > white_thresh:
        n_white += 1

back_ground_tresh = 0.25 # if x percent of the image is white then we saaay it has a white background
if (n_white / float(len(pixels))) > 0.5:
    print("white background")
else:
    print("not white background")
'''

    
text_file = open("white-non-white_Output.txt", "w")
train_path = "./data/train/"
num_img_white = 0
num_img_not_white = 0
for i in range(1,128):
    class_i_white = 0
    class_i_not_white = 0
    data_path = train_path + str(i) + "/"
    for img in os.listdir(data_path):
        filename = os.fsdecode(img)
        greyscale = Image.open( data_path + filename).convert('LA')
        
        pixels = greyscale.getdata() # get the pixels as a flattened sequence
        white_thresh = 240
        n_white = 0
        for pixel in pixels:
            if pixel[0] > white_thresh:
                n_white += 1

        back_ground_tresh = 0.25 # if x percent of the image is white then we saaay it has a white background
        if (n_white / float(len(pixels))) > back_ground_tresh:
            num_img_white += 1
            class_i_white += 1
        else:
            num_img_not_white += 1
            class_i_not_white += 1
            
    text_file.write("class: " + str(i) + " white: "+ str(class_i_white) + "\n")
    text_file.write("class: " + str(i) + " not white: " + str(class_i_not_white) + "\n")
    print("class: " + str(i) + " white: ",class_i_white)
    print("class: " + str(i) + " not white: ",class_i_not_white)

text_file.write("total whites: " + str(num_img_white) + "\n")
text_file.write("total not whites: " + str(num_img_not_white) + "\n")
text_file.close()
print("total whites: ",num_img_white)
print("total not whites: ",num_img_not_white)
```
