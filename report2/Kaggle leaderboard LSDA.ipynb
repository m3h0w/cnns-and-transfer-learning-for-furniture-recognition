```python
import pandas as pd
import numpy as np
```

```python
team_names = ["Holdet","SofaSoGood","Brick Squad",
              "K-NN or GTFO","Team Five","Class A fire",
              "space BAT","New new new pied piper","KU Group 9",
              "Realdumbteam","Oreo","LSDA-G12","LSDA13","Dream team",
              "Team ACHZ","fashion_guys","LSDA17","IKEA-CREW",
              "Brute-Force 1-NN","3d1r","LSDA-21"]
# team_names = [name.split(' ')[0] for name in team_names]
team_names
```

```python

furniture_board = pd.read_csv('./imaterialist-challenge-furniture-2018-publicleaderboard.csv')
# furniture_board = furniture_board.sort_values(['Score'])
# furniture_board = furniture_board.reset_index(drop=True)
# furniture_board.index += 1
# max_score = furniture_board.groupby(['TeamId'])['Score']
# print(max_score)
scores = []
for team in team_names:
    if(any(furniture_board['TeamName'] == team)):
        row = furniture_board[furniture_board['TeamName'] == team].head(1)
        scores.append(list(row.values[0]))
scores = np.array(scores)
print()
furniture_board_lsda = pd.DataFrame(scores, columns=list(furniture_board.columns)).sort_values(['Score'])
furniture_board_lsda
```

```python
fasion_board = pd.read_csv('./imaterialist-challenge-fashion-2018-publicleaderboard.csv')
scores = []
for team in team_names:
    if(any(fasion_board['TeamName'] == team)):
        scores.append(fasion_board[fasion_board['TeamName'] == team].tail(1).values[0])
scores = np.array(scores)
fashion_board_lsda = pd.DataFrame(scores, columns=fasion_board.columns).sort_values(['Score'], ascending=False)
fashion_board_lsda
```
